```
from __future__ import print_function 
import numpy as np
from sklearn import neighbors, datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score, recall_score

np.random.seed(7) 
breast_cancer = datasets.load_breast_cancer()
breast_cancer_X = breast_cancer.data
breast_cancer_y = breast_cancer.target

print('Labels:', np.unique(breast_cancer_y))

# split train and test 
X_train, X_test, y_train, y_test = train_test_split(
     breast_cancer_X, breast_cancer_y, test_size=50)

model = neighbors.KNeighborsClassifier(n_neighbors = 51, p = 2)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy of 1NN: %.2f %%" %(100*accuracy_score(y_test, y_pred)))
print("Precision Score: %.2f %%" %  (100*precision_score(y_test,y_pred, average='macro')))
print("Recall Score: %.2f %%" %  (100*recall_score(y_test,y_pred, average='macro')))
```
